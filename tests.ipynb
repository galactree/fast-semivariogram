```python
from fast_semivariogram import fast_semivariogram
import numpy as np
import matplotlib.pyplot as plt
```

```python
Z = np.random.normal(0,1, (200,150))

N_px = 200, 150
resolution = 0.15

box_size = resolution * np.array(N_px)
```

Functions for testing:

```python
def test_params(bins, separation_cutoff):
    params = {'bins' : bins,
    'separation_cutoff': separation_cutoff}
    return {param : param_val for param, param_val in params.items() if param_val is not None}

def plot_semivariogram(sep, svg, N):
    fig, ax = plt.subplots(1,2, figsize = (15,5))
    ax[0].plot(sep, svg, label= r'$\gamma(r)$')
    ax[0].set_xlabel(r'$r$ (pixels)')
    ax[0].set_ylabel(r'$\gamma(r)$')


    ax[1].plot(sep, N, label= r'$\gamma(r)$')
    ax[1].set_xlabel(r'$r$ (pixels)')
    ax[1].set_ylabel(r'Number of Points')
```

# Parameter Tests
1. bins is a list of evenly spaced bin edges
2. bins is a list of unevenly spaced bins
3. bins is an integer, separation_cutoff is not specified
4. bins is not specified, separation_cutoff is an integer
5. bins is an integer, separation_cutoff is an integer
6. bins is not specified, separation_cutoff is not specified

## 1. bins is a list of evenly spaced bin edges

```python
bins = np.linspace(0, 10, 100)
separation_cutoff = None
params = test_params(bins=bins, separation_cutoff=separation_cutoff)
sep, svg, N  = fast_semivariogram(Z, box_size, **params)
plot_semivariogram(sep, svg, N)
```

## 2. bins is a list of unevenly spaced bin edges

```python
bins = [0, 0.15, 0.2, 0.3, 0.45, 0.5, 0.55, 1, 1.5, 1.6, 1.75, 2]
separation_cutoff = None
params = test_params(bins=bins, separation_cutoff=separation_cutoff)
sep, svg, N  = fast_semivariogram(Z, box_size, **params)
plot_semivariogram(sep, svg, N)
```

```python
bin_sizes = np.linspace(0, 1, 20)
bins = np.cumsum(bin_sizes)
separation_cutoff = None

params = test_params(bins=bins, separation_cutoff=separation_cutoff)
sep, svg, N  = fast_semivariogram(Z, box_size, **params)
plot_semivariogram(sep, svg, N)
```

## 3. bins is an integer, separation_cutoff not specified

```python
bins = 100
separation_cutoff = None
params = test_params(bins=bins, separation_cutoff=separation_cutoff)
sep, svg, N  = fast_semivariogram(Z, box_size, **params)
plot_semivariogram(sep, svg, N)
```

## 4. bins is not specified, separation_cutoff is an integer

```python
bins = None
separation_cutoff = 10
params = test_params(bins=bins, separation_cutoff=separation_cutoff)
sep, svg, N  = fast_semivariogram(Z, box_size, **params)
plot_semivariogram(sep, svg, N)
```

## 5. bins is an integer, separation_cutoff is an integer
